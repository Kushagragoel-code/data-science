```python
import numpy as np
a = np.random.randint(-90,90,(5,5))
b = np.mean(a,axis=1)
c = np.mean(a,axis=0)


print("original matrix:\n" ,a)
print("mean row \n",b)
print("mean row \n",c)

d=a.copy()
d=np.where(d%2==0,-1,d)
print("\n")
print("replaced all even \n",d)
print("\n")
h = np.sort(a,axis=0)
i  = np.sort(a,axis=1)
print("sorted row \n",h)
print("\n")
print("sorted column \n",i)

u = np.sum(a,axis=0)
print("\n")
print("sum row \n",u)
o = np.sum(a,axis=1)
print("\n")
print("sum column \n",o)
count=0
for i in range(0,5):
  for j in range(0,5):
    if a[i][j]<0:
      count=count+1
print(count)
```

```python
import numpy as np
a = np.random.randint(1,50,(5,5))
q = np.random.randint(1,50,(5,5))
w = a.copy()
print("stock \n",a)
print()
print("demand \n",q)
count =0
ans =0
for i in range(0,5):
  for j in range(0,5):

    w[i][j]-=q[i][j]
    if(w[i][j]<0):
      count-=w[i][j]
      ans=ans+1
print()
print("remaining stock \n",w)
s=np.where(w<0,0,w)
print()
print("replace value \n",s)
print()
print("total shortage \n",count)
print()
print("product with zero shortage \n",ans)
print(np.sort(w,axis=0))
print()
print(np.sort(w,axis=1))




```
