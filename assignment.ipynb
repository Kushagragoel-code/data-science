```python

import numpy as np

n = int(input("Number of items: "))

prices = []

for i in range(n):
    prices.append(int(input()))

prices = np.array(prices)

print("Prices =", prices)

print("Total amount    =", np.sum(prices))
print("Average amount =", np.mean(prices))
print("Minimum amount =", np.min(prices))
print("Maximum amount =", np.max(prices))


print("Sorted Prices =", np.sort(prices))
print("Unique Prices =", np.unique(prices))

print("Standard Deviation       =", np.std(prices))

```
